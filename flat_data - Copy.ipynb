## Outliers Handling - Interquartile Range(IQR) Method

```python
Dataframe_4= Dataframe_3.copy()
```

The Outlier Columns are
1. floor_area_sqm
2. lease_commence_date
3. resale_price
4. storey_start
5. storey_end
6. Remaining_lease_year
7. Remaining_lease_month
Recorrect the Outliers

```python
def outlier(df,column):
  q1= df[column].quantile(0.25)
  q3= df[column].quantile(0.75)

  iqr= q3-q1

  upper_threshold= q3 + (1.5*iqr)
  lower_threshold= q1 - (1.5*iqr)

  df[column]= df[column].clip(lower_threshold, upper_threshold)
```

```python
outlier_columns=["floor_area_sqm","lease_commence_date","resale_price","storey_start","storey_end"]
for j in outlier_columns:
    outlier(Dataframe_4,j)
```

```python
for k in outlier_columns:
    plot(Dataframe_4,k)
```

```python
# Split the data into features and target
x = Dataframe_4[['Year', 'town', 'flat_type', 'floor_area_sqm', 'flat_model','storey_start', 'storey_end', 'Remaining_lease_year','Remaining_lease_month', 'lease_commence_date']]
y = Dataframe_4['resale_price']
```

Train and Test data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

linear_regression = LinearRegression()
linear_regression.fit(x_train , y_train)
y_pred = linear_regression.predict(x_test)

print(f'Mean absolute error is : {metrics.mean_absolute_error(y_test, y_pred)}')
print(f'Mean squared error is : {metrics.mean_squared_error(y_test, y_pred)}')
print(f'Root mean squared error is : {np.sqrt(metrics.mean_squared_error(y_test, y_pred))}')
print(f'R2 score is : {linear_regression.score(x_test, y_test)}')
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

decision_tree_regressor = DecisionTreeRegressor()
decision_tree_regressor.fit(x_train, y_train)
y_pred = decision_tree_regressor.predict(x_test)

print(f'Mean absolute error is : {metrics.mean_absolute_error(y_test, y_pred)}')
print(f'Mean squared error is : {metrics.mean_squared_error(y_test, y_pred)}')
print(f'Root mean squared error is : {np.sqrt(metrics.mean_squared_error(y_test, y_pred))}')
print(f'R2 score is : {decision_tree_regressor.score(x_test, y_test)}')
```
